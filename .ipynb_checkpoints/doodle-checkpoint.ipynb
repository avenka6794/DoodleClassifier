```python
import numpy as np

raw = np.load('scissors.npy')
img = raw[80:]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

for i in range(4):
    current = img[i].reshape((28,28))
    plt.imshow(current, cmap='Greys',  interpolation='none')
    plt.show()
```

```python
total = 1000

bytes = bytearray()

for index in range(total):
    current = img[index]
    for f in current:
        bytes.append(f)

newFile = open("scissors1000.bin", "wb")
newFile.write(bytes)
```
